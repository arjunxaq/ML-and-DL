A tokenizer is a tool used in natural language processing (NLP) and machine learning that breaks text into smaller pieces called tokens. These tokens are the units that models like GPT or BERT understand and work with.

What exactly does it do?
1. Splits input text:

Converts a sentence like "I love pizza!" into tokens.

Depending on the tokenizer, these might be:

Words: ["I", "love", "pizza", "!"]

Subwords: ["I", "love", "pizz", "a", "!"]

Characters: ["I", " ", "l", "o", "v", "e", " ", "p", "i", "z", "z", "a", "!"]

2. Maps tokens to numbers:

Each token is assigned an ID from a predefined vocabulary.

Example: "I" → 42, "love" → 103, "pizza" → 2045

3. Adds special tokens (if needed):

For example, <CLS> at the start or <SEP> between sentences, depending on the model.

4. Handles padding/truncation:

Ensures all input sequences are the same length for batch processing.

Why is it important?
Neural networks (like transformers) can't read raw text — they work with numbers.

Tokenizers are responsible for converting human language into something a model can process and generate predictions from.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

#initialise the tokenizer
tokenizer = AutoTokenizer.from_pretrained("siebert/sentiment-roberta-large-english")

#initialise the model
model = AutoModelForSequenceClassification.from_pretrained("siebert/sentiment-roberta-large-english")
```

The AutoTokenizer in Hugging Face Transformers is a unified interface that automatically loads the correct tokenizer class for a given pretrained model. It's like a smart wrapper that figures out whether to use BertTokenizer, GPT2Tokenizer, T5Tokenizer, etc., based on the model name or path you provide.

How does it know which tokenizer to use?
Each model checkpoint (e.g., "bert-base-uncased", "gpt2", etc.) has metadata (config.json) that tells the library which tokenizer class it requires.

##Pipeline:

In the Hugging Face Transformers library, pipeline is a high-level API that allows you to use pre-trained models for common NLP (and some vision/audio) tasks with minimal setup. It abstracts away much of the complexity of loading models and tokenizers, letting you focus on your task.

🔧 What does pipeline do?
The pipeline() function automatically:

1. Loads a pre-trained model and tokenizer suitable for the task.

2. Handles tokenization, model inference, and output formatting.

3. Makes it easy to switch between different models or tasks.



```python
from transformers import pipeline
sentiment = pipeline("sentiment-analysis")
print(sentiment("I love Hugging Face!"))
```

```python
generator = pipeline("text-generation", model="gpt2")
print(generator("Once upon a time", max_length=30))
```

```python
translator = pipeline("translation_en_to_fr")
print(translator("Hello, how are you?"))
```

```python
qa = pipeline("question-answering")

context = "Hugging Face is creating a tool that democratizes AI."
question = "What is Hugging Face creating?"

result = qa(question=question, context=context)
print(result['answer'])
```

##Other way: Load Pretrained Models and Tokenizers Manually

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

# Tokenize
inputs = tokenizer("Hugging Face is amazing!", return_tensors="pt")

# Predict
outputs = model(**inputs)
print(outputs.logits)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "sarvamai/sarvam-m"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# Add LLaMA 2 chat format
prompt = "<s>[INST] What is the capital of France? [/INST]"

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
outputs = model.generate(**inputs, max_new_tokens=100)

response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)
```
