```python
import os
import fitz  # PyMuPDF
from openai import OpenAI
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import drive
from PIL import Image
import requests
from io import BytesIO
import base64
import time

# === 1. Mount Google Drive ===
drive.mount('/content/drive')

# === 2. API Setup ===
API_KEY = "API"  # Replace with your key
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=API_KEY,
)

# === 3. Read and chunk academic book ===
def load_book_chunks(pdf_path, chunk_size=500):
    """Load and chunk PDF content"""
    try:
        print(f"Loading PDF from: {pdf_path}")
        doc = fitz.open(pdf_path)
        text = " ".join(page.get_text() for page in doc)
        doc.close()
        chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
        print(f"Successfully loaded {len(chunks)} chunks from PDF")
        return chunks
    except Exception as e:
        print(f"Error loading PDF: {e}")
        return []

# === 4. Convert local image to base64 URL ===
def image_to_base64_url(image_path):
    """Convert image to base64 URL with proper error handling"""
    try:
        # Check if file exists
        if not os.path.exists(image_path):
            raise FileNotFoundError(f"Image file not found: {image_path}")

        # Get file extension to determine MIME type
        ext = os.path.splitext(image_path)[1].lower()
        mime_type = "image/jpeg" if ext in ['.jpg', '.jpeg'] else f"image/{ext[1:]}"

        with open(image_path, "rb") as img_file:
            encoded = base64.b64encode(img_file.read()).decode()
            return f"data:{mime_type};base64,{encoded}"
    except Exception as e:
        print(f"Error converting image to base64: {e}")
        return None

# === 5. Generate caption ===
def generate_caption(image_url, context=""):
    """Generate caption with better error handling and rate limiting"""
    try:
        if not image_url:
            raise ValueError("Invalid image URL")

        messages = [
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": context if context else "Describe this image in detail."},
                    {"type": "image_url", "image_url": {"url": image_url}},
                ]
            }
        ]

        print("Sending request to API...")
        completion = client.chat.completions.create(
            model="qwen/qwen2.5-vl-72b-instruct:free",
            messages=messages,
            max_tokens=1000,  # Add explicit token limit
            temperature=0.7
        )

        result = completion.choices[0].message.content
        print(f"API response received: {len(result)} characters")
        return result

    except Exception as e:
        print(f"Error generating caption: {e}")
        return None

# === 6. Find most relevant book chunk ===
def find_relevant_chunk(chunks, query):
    """Find most relevant chunk with error handling"""
    try:
        if not chunks or not query:
            return ""

        # Filter out empty chunks
        valid_chunks = [chunk for chunk in chunks if chunk.strip()]
        if not valid_chunks:
            return ""

        vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
        X = vectorizer.fit_transform(valid_chunks + [query])
        similarities = cosine_similarity(X[-1], X[:-1])[0]
        best_idx = np.argmax(similarities)

        print(f"Found relevant chunk with similarity: {similarities[best_idx]:.3f}")
        return valid_chunks[best_idx]

    except Exception as e:
        print(f"Error finding relevant chunk: {e}")
        return ""

# === 7. Compose caption with context ===
def caption_with_context(image_url, context_chunk):
    """Generate detailed caption with context"""
    prompt = '''You are an expert AI model specialized in generating highly detailed, precise, and educationally relevant captions for images specifically tailored to the educational technology domain. Your task is to create captions that thoroughly capture and articulate the full range of visual elements, underlying educational concepts, and contextual significance presented in each image.

When provided with an image description or context, follow these steps:

1. Analyze the visual content carefully, identifying and naming all key objects, subjects, and scenes, especially those pertinent to educational themes and settings.
2. Elaborate on the educational context and significance of the elements within the image, explaining their role in the learning process or educational concept being conveyed.
3. Use rich, precise, and contextual language to amplify understanding, ensuring the caption is detailed enough to support effective text embedding generation for hybrid vector searches.
4. Before composing the caption, reason methodically through the image's contents and concepts to guarantee accuracy, completeness, and relevance.
5. Ensure the caption reflects pedagogical intent, highlighting relationships, activities, or interactions that promote learning.

Maintain a balance between descriptive richness and conciseness, avoiding lists or bullet points.

Return the caption as a single, coherent paragraph without formatting. It should be sufficiently detailed to assist in accurate image retrieval and semantic understanding within educational technology applications.'''

    full_context = prompt
    if context_chunk:
        full_context += f"\n\nRelevant context from textbook:\n{context_chunk}"

    return generate_caption(image_url, context=full_context)

# === 8. Main Paths ===
PDF_PATH = "/content/drive/MyDrive/Dataset/Books/AI.pdf"
IMAGE_DIR = "/content/drive/MyDrive/Artificial_Intelligence"
OUTPUT_DIR = "/content/drive/MyDrive/Dataset/descriptions/Artificial_intelligence"

# === 9. Pre-flight checks ===
print("=== PRE-FLIGHT CHECKS ===")
print(f"PDF Path exists: {os.path.exists(PDF_PATH)}")
print(f"Image Directory exists: {os.path.exists(IMAGE_DIR)}")
if os.path.exists(IMAGE_DIR):
    image_files = [f for f in os.listdir(IMAGE_DIR) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
    print(f"Found {len(image_files)} image files")
    if image_files:
        print(f"Sample files: {image_files[:3]}")
else:
    print("Image directory not found!")
    exit(1)

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"Output directory: {OUTPUT_DIR}")

# === 10. Load book once ===
print("\n=== LOADING BOOK ===")
chunks = load_book_chunks(PDF_PATH)
if not chunks:
    print("Warning: No chunks loaded from PDF. Proceeding without context.")

# === 11. Process all images with better error handling ===
print("\n=== PROCESSING IMAGES ===")
successful_count = 0
failed_count = 0

for i, image_file in enumerate(image_files):
    image_path = os.path.join(IMAGE_DIR, image_file)
    output_file = os.path.join(OUTPUT_DIR, f"{os.path.splitext(image_file)[0]}.txt")

    # Skip if already processed
    if os.path.exists(output_file):
        print(f"Skipping {image_file} (already processed)")
        continue

    print(f"\n[{i+1}/{len(image_files)}] Processing: {image_file}")

    try:
        # Step 1: Convert image to base64
        print("  - Converting image to base64...")
        image_url = image_to_base64_url(image_path)
        if not image_url:
            raise Exception("Failed to convert image to base64")

        # Step 2: Generate initial caption
        print("  - Generating initial caption...")
        pre_caption = generate_caption(image_url)
        if not pre_caption:
            raise Exception("Failed to generate initial caption")

        # Step 3: Find relevant context
        print("  - Finding relevant context...")
        context_chunk = find_relevant_chunk(chunks, pre_caption)

        # Step 4: Generate final caption with context
        print("  - Generating final caption with context...")
        final_caption = caption_with_context(image_url, context_chunk)
        if not final_caption:
            raise Exception("Failed to generate final caption")

        # Step 5: Save caption
        print("  - Saving caption...")
        with open(output_file, "w", encoding='utf-8') as f:
            f.write(final_caption)

        print(f"  ✓ Success! Caption saved to {os.path.basename(output_file)}")
        successful_count += 1

        # Rate limiting to avoid API issues
        time.sleep(1)

    except Exception as e:
        print(f"  ✗ Error with {image_file}: {e}")
        failed_count += 1

        # Save error log
        error_file = os.path.join(OUTPUT_DIR, f"{os.path.splitext(image_file)[0]}_ERROR.txt")
        with open(error_file, "w") as f:
            f.write(f"Error processing {image_file}: {str(e)}")

print(f"\n=== PROCESSING COMPLETE ===")
print(f"Successfully processed: {successful_count}")
print(f"Failed: {failed_count}")
print(f"Total files: {len(image_files)}")
```
